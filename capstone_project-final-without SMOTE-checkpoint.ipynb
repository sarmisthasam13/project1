```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lrtenure,urtenure=remove_outlier(data['Tenure'])
data['Tenure']=np.where(data['Tenure']>urtenure,urtenure,data['Tenure'])

lrtenure,urtenure=remove_outlier(data['Tenure'])
data['Tenure']=np.where(data['Tenure']<lrtenure,lrtenure,data['Tenure'])


lrwh,urwh=remove_outlier(data['WarehouseToHome'])
data['WarehouseToHome']=np.where(data['WarehouseToHome']>urwh,urwh,data['WarehouseToHome'])
lrwh,urwh=remove_outlier(data['WarehouseToHome'])
data['WarehouseToHome']=np.where(data['WarehouseToHome']<lrwh,lrwh,data['WarehouseToHome'])


lrhr,urhr=remove_outlier(data['HourSpendOnApp'])
data['HourSpendOnApp']=np.where(data['HourSpendOnApp']>urhr,urhr,data['HourSpendOnApp'])
lrhr,urhr=remove_outlier(data['HourSpendOnApp'])
data['HourSpendOnApp']=np.where(data['HourSpendOnApp']<lrhr,lrhr,data['HourSpendOnApp'])


lror,uror=remove_outlier(data['OrderAmountHikeFromlastYear'])
data['OrderAmountHikeFromlastYear']=np.where(data['OrderAmountHikeFromlastYear']>uror,uror,data['OrderAmountHikeFromlastYear'])
lror,uror=remove_outlier(data['OrderAmountHikeFromlastYear'])
data['OrderAmountHikeFromlastYear']=np.where(data['OrderAmountHikeFromlastYear']<lror,lror,data['OrderAmountHikeFromlastYear'])


lrcp,urcp=remove_outlier(data['CouponUsed'])
data['CouponUsed']=np.where(data['CouponUsed']>urcp,urcp,data['CouponUsed'])
lrcp,urcp=remove_outlier(data['CouponUsed'])
data['CouponUsed']=np.where(data['CouponUsed']<lrcp,lrcp,data['CouponUsed'])


lroc,uroc=remove_outlier(data['OrderCount'])
data['OrderCount']=np.where(data['OrderCount']>uroc,uroc,data['OrderCount'])
lroc,uroc=remove_outlier(data['OrderCount'])
data['OrderCount']=np.where(data['OrderCount']<lroc,lroc,data['OrderCount'])


lrdo,urdo=remove_outlier(data['DaySinceLastOrder'])
data['DaySinceLastOrder']=np.where(data['DaySinceLastOrder']>urdo,urdo,data['DaySinceLastOrder'])
lrdo,urdo=remove_outlier(data['DaySinceLastOrder'])
data['DaySinceLastOrder']=np.where(data['DaySinceLastOrder']<lrdo,lrdo,data['DaySinceLastOrder'])


lrca,urca=remove_outlier(data['CashbackAmount'])
data['CashbackAmount']=np.where(data['CashbackAmount']>urca,urca,data['CashbackAmount'])

lrca,urca=remove_outlier(data['CashbackAmount'])
data['CashbackAmount']=np.where(data['CashbackAmount']<lrca,lrca,data['CashbackAmount'])

```
